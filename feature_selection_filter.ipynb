### Data transformation

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
#import data
reviews = pd.read_csv('reviews.csv')
```

```python
#print column names
reviews.columns
```

```python
reviews.info()
```

```python
#look at the counts of recommended
reviews.recommended.value_counts()
```

```python
#create binary dictionary
binary_dict = {True: 1, False: 0}

#transform column
reviews["recommended_trans"] = reviews.recommended.map(binary_dict)
```

```python
reviews.recommended_trans.value_counts()
```

```python
#look at the counts of rating
reviews.rating.value_counts()
```

```python
# create dictionary
rating_dict = {'Loved it': 5, 'Liked it': 4, 'Was okay': 3, 'Not great': 2, 'Hated it': 1}

# transform rating column
reviews['rating_trans'] = reviews.rating.map(rating_dict)
```

```python
# print transformed column values
reviews.rating_trans.value_counts()
```

```python
#get the number of categories in a feature
reviews.department_name.value_counts()
```

```python
# perform get_dummies
one_hot = pd.get_dummies(reviews.department_name)
```

```python
# join the new columns back onto the original
reviews = reviews.join(one_hot)
```

```python
# print column names
reviews.columns
```

```python
# transform review_date to date-time data
reviews['review_date_trans'] = pd.to_datetime(reviews.review_date)
```

```python
# print review_date data type 
print(reviews.review_date_trans.dtype)
```

```python
# get numerical columns
reviews = reviews[['clothing_id', 'age', 'recommended_trans', 'rating_trans', 'Bottoms', 'Dresses', 'Intimate', 'Jackets', 'Tops', 'Trend']].copy()
```

```python
# reset index
reviews = reviews.set_index('clothing_id')
```

```python
# instantiate standard scaler
scaler = StandardScaler()
```

```python
# fit transform data
scaler.fit_transform(reviews)
```
