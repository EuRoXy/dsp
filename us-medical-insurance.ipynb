# U.S. Medical Insurance

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
# create the initial variables below
age = df['age']
sex = df['sex']
bmi = df['bmi']
children = df['children']
smoker = df['smoker']
region = df['region']
charges = df['charges']
charges
```

```python
class PatientInfo:
    def __init__(self, age, sex, bmi, children, smoker, region, charges):
        self.age = age
        self.sex = sex
        self.bmi = bmi
        self.children = children
        self.smoker = smoker
        self.region = region
        self.charges = charges

    def analyze_age(self):
        age_mn = self.age.mean()
        age_mn_rd = np.round(age_mn, 1)
        print("Average patient age:", age_mn_rd)

    def analyze_sex(self):
        males = self.sex.value_counts()[0]
        females = self.sex.value_counts()[1]        
        print("Count for male:", males)
        print("Count for female:", females)

    def unique_regions(self):
        print(self.region.value_counts())
    
    def avg_charges(self):
        charge_mn = self.charges.mean()
        charge_mn_rd = np.round(charge_mn, 2)
        print("Average yearly medical insurance charges:", charge_mn_rd)
```

```python
patient_info = PatientInfo(age, sex, bmi, children, smoker, region, charges)
```

```python
patient_info.analyze_age()
```

```python
patient_info.analyze_sex()
```

```python
patient_info.unique_regions()
```

```python
patient_info.avg_charges()
```
